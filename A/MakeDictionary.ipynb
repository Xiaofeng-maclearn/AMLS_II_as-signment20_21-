# Dictionary Generation

In this python, we generate a dictionary

```python
# import libraries
import numpy as np
import pandas as pd
import re
```

```python
# open and read the processed twitter data
f = open("processed_twitter.txt","r",encoding='utf-8')

result=[]
for line in f.readlines():
            line=line.strip().split("\t")
            result.append(line)
f.close()

df=pd.DataFrame(result[1:]).dropna(axis=0)
df.columns = ['processed','sent']
print(df)
```

```python
# Generate the dictionary list
d = {}
for i in range(df.shape[0]):
    listOfTokens = re.split(r'\W',df.iloc[i,0])
    for j in range(len(listOfTokens)):
        token = listOfTokens[j]
        # short words are excluded
        if(len(token)<3):
            continue
        if token in d:
            d[token] = d[token]+1
        else:
            d[token] = 1
print(d)
```

```python
# generate a dataframe based on the dictionary list, so that we can save it into a file
df2 = pd.DataFrame(d.items())
df2.columns = ['word','frequency']
```

```python
print(df2)
```

```python
# save the dictionary
df2.sort_values(['frequency'],ascending=False).to_csv('dictionary.txt',sep = '\t',index=False)
```
